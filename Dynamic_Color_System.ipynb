```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img=cv2.imread('1.jpeg')

img.shape
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img)
plt.show()
```

```python
#flatten into linear Array
all_pixels=img.reshape(-1,3)
print(all_pixels)
#shape
print(all_pixels.shape)
```

```python
DominantColors=5
km=KMeans(n_clusters=DominantColors)
km.fit(all_pixels)
```

```python
centers=km.cluster_centers_
centers
```

```python
centers=np.array(centers,dtype='uint8')
centers
```

```python
arr_rgb=centers.tolist()
print(arr_rgb)
```

```python
i=1
plt.figure(0,figsize=(8,2))
colors=[]
for x in centers:
  plt.subplot(1,5,i)
  plt.axis(False)
  i+=1
  colors.append(x)
  a=np.zeros((100,100,3),dtype='uint8')
  a[:,:,:]=x
  plt.imshow(a)
plt.show
```

**Convert The RGB Image Into HSV**

```python
hsvimg = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
hsvimg.shape
```

```python
plt.figure(figsize=(10,5))
plt.imshow(hsvimg)
plt.show()
```

```python
#flatten into linear Array
all_pixels_hsv=hsvimg.reshape(-1,3)
print(all_pixels_hsv)
#shape
print(all_pixels_hsv.shape)
```

```python
DominantColors=5
kms=KMeans(n_clusters=DominantColors)
kms.fit(all_pixels_hsv)
```

```python
centers_hsv=kms.cluster_centers_
centers_hsv
```

```python
centers_hsv=np.array(centers_hsv,dtype='uint8')
centers_hsv
```
